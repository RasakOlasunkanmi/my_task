```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer("all-MiniLM-L6-v2")

sentences = [
    "The dog is playing in the park",
    "A puppy is running outside",
    "The cat is sleeping on the couch",
    "Python is a programming language",
    "Machine learning models need data",
    "I love coding in Python"
]

# Generate embeddings for all sentences
embeddings = model.encode(sentences)

# Compare Sentence 1 and Sentence 4 to all others
def compare(index):
    print(f"\nComparing to: '{sentences[index]}'\n")
    for i, sentence in enumerate(sentences):
        sim = cosine_similarity([embeddings[index]], [embeddings[i]])[0][0]
        print(f"Similarity to: '{sentence}'")
        print(f"Score: {sim:.3f}\n")

compare(0)  # Sentence 1
compare(3)  # Sentence 4
```

Similarity Analysis:

Query: "The dog is playing in the park"

Most similar: "A puppy is running outside"

Least similar: "Python is a programming language"

Observations: Animal-related sentences group together; coding topics show near-zero similarity.

Query: "Python is a programming language"

Most similar: "I love coding in Python"

Least similar: "A puppy is running outside" / "The cat is sleeping on the couch"

Observations: Programming-related sentences cluster strongly; animal sentences score very low.

Recommended similarity threshold: 0.30
