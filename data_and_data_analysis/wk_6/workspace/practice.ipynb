```python
#1. Total number of students

Total_of_students = df["total_students"].sum()
Total_of_students
```

```python
#2. Total annual family income

Total_annual_family_income = df["annual_family_income"].sum()
Total_annual_family_income
```

```python
#3. Total colleges

Total_colleges = df["college_id"].count()
Total_colleges
```

```python
#4. Average CGPA

Average_cgpa = round(df["cgpa"].mean(), 2)
Average_cgpa
```

```python
#5. Total Placement Rate

Total_placement_rate = df["placement_rate"].sum()
Total_placement_rate
```

```python
df['placement_rate_percentage'] = (df['placement_rate'] / Total_placement_rate) * 100
```

```python
df
```

```python
# Average placement rate
Average_placement_rate_percentage = df["placement_rate_percentage"].mean()
Average_placement_rate_percentage
```

```python
# Average annual family income
Average_annual_family_income = df["annual_family_income"].mean()
Average_annual_family_income
```

```python
# Total faculty count
Total_faculty_count = df["faculty_count"].sum()
Total_faculty_count
```

```python
#6. Average faculty count per college
Average_faculty_count_per_college = round(Total_faculty_count / Total_colleges, 2)
Average_faculty_count_per_college
```

```python
#7. Total research papers published

Total_research_papers_published = df["research_papers_published"].sum()
Total_research_papers_published
```
