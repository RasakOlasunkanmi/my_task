Computing mean, median, mode, variance, and standard deviation for all the numeric columns:

age (numeric, though may be categorical-like — but still can compute stats)
hdlngth (head length)
skullw (skull width)
totlngth (total length)
taill (tail length)

eye (eye size)
chest (chest girth)
belly (belly girth)

These are continuous morphometric measurements (and age as ordinal/continuous). They are ideal for center & spread measures.

Note:
Continuous vs. unlimited
The key takeaway is that "continuous" refers to the type of measurement, not to the duration of the process.
Continuous measurement: The measurements for skull width or eye size can, in theory, take any value within a range. For example, a skull might be 150.1 millimeters wide or 150.2 millimeters wide. You can use decimals and fractions.

Unlimited growth: The body's biological and developmental processes impose upper limits on how much these structures can grow. Normal, healthy growth is continuous during a specific period but not infinite.

```python
# Measures of Centre and Spread for numeric columns

# Select numeric columns
num_cols = ["age", "hdlngth", "skullw", "totlngth", "taill", "footlgth", "earconch", "eye", "chest", "belly"]
```

```python
# Compute stats
stats1 = pd.DataFrame({
    "Mean": df[num_cols].mean().round(2),
    "Median": df[num_cols].median().round(2),
    "Mode": df[num_cols].mode().iloc[0].round(2),      # mode may return multiple, take first
    "Variance": df[num_cols].var().round(2),
    "Std Dev": df[num_cols].std().round(2)
})

stats1
```

### Interpretation of Measures of Centre and Spread

- **Age**:  
  The mean and median are close, suggesting the distribution is fairly symmetric.  
  Variance and standard deviation indicate the spread of age classes.  
  Mode shows the most common age class in the dataset.  

- **Head Length (hdlngth)**:  
  The mean vs. median comparison shows whether the distribution is skewed.  
  Variance and standard deviation reflect how much head length varies across possums.  

- **Skull Width (skullw)**:  
  Interpretation similar to above – check if mean ≈ median for symmetry.  
  Larger variance indicates more diversity in skull width across the sample.  

- **Total Length (totlngth)**:  
  Compare mean, median, and mode for skewness.  
  Spread shows overall body length variation.  

- **Tail Length (taill)**:  
  Consider if the distribution is symmetric or skewed.  
  Spread values will show if there is consistency in tail size.  

- **Foot Length (footlgth)**:  
  Note if the mode matches the median/mean — that suggests normal-like distribution.  

- **Ear Length (earconch)**, **Eye Size (eye)**, **Chest**, **Belly**:  
  Similar interpretations: look at how close mean and median are (symmetry),  
  and whether variance/std dev are high (wide spread) or low (consistent measurements).  

**NOTE**

**General Insight:**  
- If **mean ≈ median ≈ mode**, the variable is roughly symmetric (normal-like).  
- If **mean > median**, distribution is right-skewed; if **mean < median**, left-skewed.  
- Higher **variance / std dev** = more spread in the data.  
- The mode helps identify the most common measurement value.  


```python
stats2 = pd.DataFrame({
    "Maximum": df[num_cols].max().round(2),
    "Minimum": df[num_cols].min().round(2),
    "Range": (np.subtract(df[num_cols].max(), df[num_cols].min())).round(2),
    "25th %ile": df[num_cols].quantile(0.25).round(2),
    "50th %ile": df[num_cols].quantile(0.50).round(2),
    "75th %ile": df[num_cols].quantile(0.75).round(2),
    "IQR": (np.subtract(df[num_cols].quantile(0.75), df[num_cols].quantile(0.25))).round(2),
    "Skewness": df[num_cols].skew().round(2),
    "Kurtosis": df[num_cols].kurtosis().round(2)
})

stats2
```

**NOTE**
- Maximum / Minimum / Range: absolute span of observed values.
- 25th / 50th / 75th percentile (Q1 / median / Q3): where the middle 50% sits.
- IQR = Q3 − Q1: spread of the middle 50% (robust measure of variability).
- Skewness: sign and magnitude show tail direction (≈0 symmetric, >0 right/skewed, <0 left/skewed).
- Kurtosis: relative tail/peak: positive → heavier tails/peaky; negative → lighter tails/flat-topped.


```python
# Calculate Q1, Q3, and IQR
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

print("IQR:\n", IQR)
```

**age**

- Range 1 -> 9 (range = 8). Middle 50% from 2.25 to 5.00 (IQR = 2.75). Median = 3.
- Skewness 0.54: mild right skew (a few older animals).
- Kurtosis −0.28: slightly lighter tails than a normal distribution (but near 0 — not extreme).

**Interpretation:** age classes are concentrated in low values (young animals), with a handful of higher ages. Use median/IQR to summarize age; tests that assume normality may be only roughly applicable.

- Range 50.0 -> 68.6 (span 18.6). Middle 50% = 54.98 -> 58.10 (IQR = 3.12). Median ≈ 56.35.

- Kurtosis 2.48: substantially positive — heavier tails / more extreme values than normal.

**Interpretation:** skull width shows real asymmetry and heavier tails — expect several large skull widths (outliers). Use median/IQR or robust tests when comparing groups; log transform may reduce skew.

- Kurtosis −0.10: very close to normal.

**Interpretation:** total length varies more (IQR = 6) than some other measures — good candidate as a size proxy (PC1 or size covariate in regressions).

- Kurtosis 0.40: slight heavy-tailed tendency but modest.

**Interpretation:** tail length middle spread small (IQR 2.12) relative to range — most animals have similar tails, with a couple extremes (we saw both min and max beyond IQR fences earlier).

- Range 60.3 -> 77.9 (span 17.6). Middle 50% = 64.60 -> 72.50 (IQR = 7.90). Median = 68.

**Interpretation:** ear length middle 50% is relatively wide — individuals vary, but few extreme tails. Use mean/SD carefully; median/IQR robust.

**chest**

**NOTE**

**Cross-variable observations (big picture)**

- Most morphometrics are roughly symmetric (skewness close to 0) and have modest kurtosis — meaning distributions are generally well-behaved for parametric tests except skullw (clear right skew and heavier tails) and age (mild right skew).

- IQR relative to range: totlngth and footlgth have larger IQRs (6.0 and 7.9) — more variability in those traits. eye and chest have small IQRs (1.32, 2.50) — very consistent measures.

- Skull width stands out: highest positive skew and positive kurtosis — expect several extreme large skull widths; use robust stats or transform before parametric tests.

- Foot and ear lengths are platykurtic (negative kurtosis) — wider middle, fewer extreme tails.

```python
# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outlier_def = (df[num_cols] < lower_bound) | (df[num_cols] > upper_bound)
```

```python
# Count of outliers per variable
display(outlier_def.sum().to_frame("IQR_outliers"))   # OR: print("IQR_outlier:\n", outlier_def.sum())
```

```python
flagged_iqr = df.loc[outlier_def.any(axis=1)].copy()
flagged_iqr["outlier_in"] = outlier_def[outlier_def.any(axis=1)].astype(int).dot(outlier_def.columns + ", ").str.rstrip(", ")
flagged_iqr
```

```python
# Rows flagged in any variable
flagged_iqr = df.loc[outlier_def.any(axis=1)].copy()

# Add a column listing which variables caused the flag
flagged_iqr["flagged_vars"] = outlier_def[outlier_def.any(axis=1)].apply(
    lambda row: ",".join(row.index[row].tolist()), axis=1
)

flagged_iqr
```

```python
# This shows a count of how many times each variable produced an outlier
flagged_iqr["flagged_vars"].str.split(",").explode().value_counts()
```
