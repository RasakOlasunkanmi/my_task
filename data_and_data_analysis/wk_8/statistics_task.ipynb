Task1– Mean, Median & Mode
```c
1: Class Test Scores
The following are the Mathematics test scores (out of 20) for a class of 15 students:
 5, 8, 10, 12, 12, 12, 13, 14, 14, 15, 16, 18, 18, 20, 20
Find the mean, median, and mode of the scores.


If the lowest score (5) was wrongly recorded and should have been 10, how would the mean change? Would the median or mode change? Explain.

Solution:

1. Class Test Scores
```c
Scores (out of 20): 
5, 8, 10, 12, 12, 12, 13, 14, 14, 15, 16, 18, 18, 20, 20  

### Mean
Mean = {5 + 8 + 10 + 12 + 12 + 12 + 13 + 14 + 14 + 15 + 16 + 18 + 18 + 20 + 20}/{15}

Mean = 13.8

Mean = 13.8

### Median
- Since there are 15 scores (odd number), the median is the 8th score when arranged in order.
- Ordered data is already given.  
8th score = 14

Median = 14

### Mode
- The most frequent value(s).  
12 appears 3 times, 14 appears 2 times, 18 appears 2 times, 20 appears 2 times.  
So, 
Mode = 12


### If lowest score (5) was wrongly recorded as 10:
New dataset:  
10, 8, 10, 12, 12, 12, 13, 14, 14, 15, 16, 18, 18, 20, 20  

- New sum = 207 − 5 + 10 = 212  
- New mean = 212 ÷ 15 = 14.13 

Mean increases from 13.8 to 14.13  

Median:
- Still the 8th value = 14 (no change).  

Mode:
- 12 still occurs most frequently. 
Mode = 12 (unchanged).

2: Daily Commute Times
```c
A group of 9 students reported the time (in minutes) it takes them to get to school each day:
 15, 20, 20, 25, 25, 30, 60, 65, 120
Find the mean, median, and mode of the data.


Which measure (mean, median, or mode) do you think best represents the “typical” time it takes students to get to school? Why?
 (Hint: Think about the effect of the student who takes 120 minutes.)


#### Solution
2.  Daily Commute Times
```c
Times (in minutes): 
15, 20, 20, 25, 25, 30, 60, 65, 120  

Mean
Mean = {15 + 20 + 20 + 25 + 25 + 30 + 60 + 65 + 120} / {9}
Mean = {380} / {9}

Mean = 42.2 minutes


Median
- 9 student's data with middle value as 5th value = 25  
5th value = 25

Median = 25 minutes


Mode
- 20 appears twice, 25 appears twice.  
Mode = 20 and 25 (bimodal).

Best Measure of Typical Time
- The mean (42.2) is affected by the heavy value (120 minutes).  
- The median (25 minutes) is a better representation of the “typical” time, since most students’ times cluster around 20–30 minutes.  

Answer: Median best represents the typical time.

3: Shoe Sizes
```c
The shoe sizes of students in a class are:
 38, 39, 38, 40, 41, 42, 39, 39, 40, 38, 37, 39, 40, 41, 42
Find the mode of the data.


If a shoe company wants to donate free shoes, which size should they produce most of? Why?


What if they only want to produce one size that would fit the "average" student? Should they use mean, median, or mode? Defend your answer.

#### Solution
```c
3. Shoe Sizes

Data:  
38, 39, 38, 40, 41, 42, 39, 39, 40, 38, 37, 39, 40, 41, 42  

Mode
- 39 occurs 4 times  
- 38 occurs 3 times  
- 40 occurs 3 times  
So, 
Mode = 39


Which size should the company produce most of?
- The mode (39) is the most common size.  
- To maximize correct fit, the company should produce more size 39 shoes.


If they want to produce only one size for the "average" student:
- The 'mean' might give a non-integer (e.g., 39.5) which isn’t a valid shoe size.  
- The 'median' gives a central size (though, 39) but might still not represent majority fit.  
- The 'mode' (39) best captures the most frequently worn size.  

Answer: Use the mode (size 39), since it ensures the largest number of students can actually wear the shoes.
